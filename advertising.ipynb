```python
import pandas as pd
```

```python
sale=pd.read_csv("Advertising.csv")
print(sale.head())
print(sale.describe())
```

```python
import seaborn as sns
sale=pd.DataFrame(sale)
```

```python
sns.displot(sale["Newspaper"])
```

```python
sns.displot(sale["Radio"])
```

```python
sns.displot(sale["TV"])
```

```python
sns.pairplot(sale,x_vars=["TV","Radio","Newspaper"],y_vars=["Sales"],kind="scatter",height=5)
```

```python
x=sale[["Newspaper","TV","Radio"]]
y=sale["Sales"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5)

```

<h1>Linear Regrission

```python
from sklearn.linear_model import LinearRegression
z=LinearRegression()
z.fit(x_train,y_train)
```

```python
pre1=z.predict(x_test)
print(pre1)
```

```python
from sklearn.metrics import r2_score
error=r2_score(y_test,pre1)
print((error)*100)
```

<h1>Random forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
i=RandomForestRegressor()
i.fit(x_train,y_train)
```

```python
pre2=i.predict(x_test)
print(pre2)
```

```python
from sklearn.metrics import r2_score
error2=r2_score(y_test,pre2)
print((error2)*100)
```

```python
import numpy as np
input_str = input(" Enter sales value for 'TV','Newspaper','Radio' ")

# Convert the input string into a list of integers
input_list = [int(x) for x in input_str.split()]

# Print the array
print("Array:", input_list)
x_new = np.array(input_list).reshape(1, -1)
prediction=i.predict(x_new)
print("Predicted Sales Values: {}".format(prediction))
```
